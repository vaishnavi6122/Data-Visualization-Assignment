```python
x=np.arange(1,11)
y=3*x+5
plt.title('line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y,'go--')
```

```python
##sine waveform

x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('sine waveform')

plt.plot(x,y,color='g')
plt.show()
```

```python
x=np.arange(0,10)
y=x*x
plt.plot(x,y)
```

```python
plt.plot(x,y,'r*',linestyle='dashed')
```

```python
##creating subplot
##multiple plots in single plane

plt.subplot(2,2,1)
plt.plot(x,y,'go')
plt.subplot(2,2,2)
plt.plot(x,y,'r')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'b*--')



```
